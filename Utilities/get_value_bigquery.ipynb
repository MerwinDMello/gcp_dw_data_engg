```python
from google.cloud import bigquery
import google.auth
import pandas as pd
from datetime import datetime
```

```python
import os
os.environ["HTTP_PROXY"] = "proxy.nas.medcity.net:80"
os.environ["HTTPS_PROXY"] = "proxy.nas.medcity.net:80"
```

```python
credentials, project_id = google.auth.default(scopes=["https://www.googleapis.com/auth/cloud-platform"])
print(project_id)
bq_client = bigquery.Client(credentials=credentials, project=project_id)
```

```python
 
def lookup_bq_data():
    dt1 = datetime.now()
    try:
        table_name = 'ecw_reportsecuritypbiusercoid'
        target_query = "SELECT last_load_ctrl_timestamp FROM edwpsc_ac.load_ctrl_timestamp WHERE table_name = '{}'".format(table_name)
        target_rs = bq_client.query(target_query, project=project_id, location='US').result()
        for target_row in target_rs:
            print(target_row.last_load_ctrl_timestamp)

        dt2 = datetime.now()
        print(dt2-dt1)

    except Exception as e1:
        print(e1)
        pass
```
