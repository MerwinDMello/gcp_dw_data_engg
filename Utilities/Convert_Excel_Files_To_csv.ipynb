```python
import os
os.environ["HTTP_PROXY"] = "proxy.nas.medcity.net:80"
os.environ["HTTPS_PROXY"] = "proxy.nas.medcity.net:80"
```

```python
import pandas as pd
```

```python
excel_files_path_folder = r"InputFiles\Excel_Files" 
csv_files_path_folder = r"OutputFiles\Convert_csvs"
input_file_extension = ".xlsx"
output_file_extension = ".csv"
```

```python
print(excel_files_path_folder)
print(csv_files_path_folder)
```

```python
file_list = [filename for filename in os.listdir(f"{excel_files_path_folder}") if filename.endswith(input_file_extension)]
for filename in file_list:
    filename_without_ext = filename.replace(input_file_extension, '')
    source_file_path = os.path.join(excel_files_path_folder, filename)
    read_file = pd.read_excel(source_file_path, sheet_name="RAWDATA")
    target_file_path = os.path.join(csv_files_path_folder, filename_without_ext + output_file_extension)
    print(target_file_path)
    read_file.to_csv (target_file_path, index = None, header=True, sep=',', quotechar='"', encoding="utf-8")
```
