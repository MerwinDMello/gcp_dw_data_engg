```python
import requests
from requests.auth import HTTPBasicAuth

# Replace these variables with your Jira instance URL, username, and API token
jira_url = 'https://merwindmello.atlassian.net/'
username = 'merwin.dmello@gmail.com'
api_token = ""
# Define the JQL query to retrieve issues of type "Epic"
jql_query = 'issuetype=Epic'

# Set up the API endpoint and parameters
api_endpoint = f'{jira_url}/rest/api/3/search'
params = {
    'jql': jql_query,
    'maxResults': 100,  # Adjust the number of results as needed
    'fields': 'summary,issuetype,description,status'
}

# Make the API request
response = requests.get(api_endpoint, auth=HTTPBasicAuth(username, api_token), params=params)

print(response.json())

# Check if the request was successful
if response.status_code == 200:
    issues = response.json()['issues']
    for issue in issues:
        print(f"Status: {issue['key']}")
        print(f"Summary: {issue['fields']['summary']}")
        print(f"Summary: {issue['fields']['issuetype']['name']}")
        print(f"Description: {issue['fields']['description']}")
        print(f"Status: {issue['fields']['status']['name']}") 
        print('-' * 40)
else:
    print(f"Failed to retrieve issues: {response.status_code} - {response.text}") 
```
