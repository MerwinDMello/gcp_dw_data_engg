```python
import yaml
import json
```

```python
datatype_conv_config = r"InputFiles\DataType_Conversion.yaml"
```

```python
with open(datatype_conv_config, 'r') as datatype_conv_conf:
    datatype_conv_yaml = yaml.safe_load(datatype_conv_conf)
    datatype_conv_json = json.loads(json.dumps(datatype_conv_yaml, default=str))
```

```python
print(datatype_conv_json["conversion"])
```

```python
datatype_conv_config_item = next(filter(lambda item: item['source_dialect'] == "sql_server", datatype_conv_json["conversion"]),None)
```

```python
print(datatype_conv_config_item['datatypes'])
```

```python
datatype_conv_config_datatype_item = filter(lambda item: item['source_datatype'] == "varchar", datatype_conv_config_item['datatypes'])
```
