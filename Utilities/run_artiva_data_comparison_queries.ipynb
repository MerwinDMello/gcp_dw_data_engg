```python
%%bigquery
-- Record Count for core table collection_agency_placement_detail 
SELECT
COUNT(*) as Rec_Count,
Max(dw_last_update_date_time) as dw_last_update_date_time
FROM
  `hca-hin-qa-cur-parallon.edwpbs_base_views.collection_agency_placement_detail`
WHERE artiva_instance_code = 'AR1';
```

```python
%%bigquery
-- Record Count for mirrored table collection_agency_placement_detail 
SELECT
COUNT(*) as Rec_Count,
Max(dw_last_update_date_time) as dw_last_update_date_time
FROM
  `hca-hin-qa-cur-parallon.edwpbs_base_views_copy.collection_agency_placement_detail`
WHERE artiva_instance_code = 'AR1';
```

```python
%%bigquery
-- Data Comparison for collection_agency_placement_detail table
SELECT
* EXCEPT(dw_last_update_date_time, placement_sid)
FROM `hca-hin-qa-cur-parallon.edwpbs.collection_agency_placement_detail`
WHERE artiva_instance_code = 'AR1'
EXCEPT DISTINCT
SELECT
* EXCEPT(dw_last_update_date_time, placement_sid)
FROM `hca-hin-qa-cur-parallon.edwpbs_base_views_copy.collection_agency_placement_detail`
WHERE artiva_instance_code = 'AR1';
```
