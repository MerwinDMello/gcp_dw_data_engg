```python
from datetime import datetime, timedelta
import os
from jinja2 import Environment, FileSystemLoader
```

```python
def read_contents_file(file_path, encoding_scheme='cp1252'):
    inputfile = open(file_path, 'r', encoding=encoding_scheme)
    return inputfile.read()
```

```python
#Write file to local directory
def write_file_local(path,file_data):
    
    with open(path, 'w') as file:
        file_string = file_data
        file.write(file_string)
```

```python
template_folder = "InputFiles\Template"
template_name = "script_template_loop"
jinja_extension = ".j2"
yaml_extension = ".yaml"
output_path_folder = f"OutputFiles\Script"
output_file_name = "Test_Loop"

name_list = [{"name":"Merwin"},{"name":"Melwin"},{"name":"Merlin"}]

script_dict = {
    "name_list":name_list
}
```

```python
#get the DDLs for the tables from teradata 
def gen_script_using_template():
    dt1 = datetime.now()

    try:
        if not os.path.exists(output_path_folder):
            os.makedirs(output_path_folder)

        # Jinja2 Environment
        env = Environment(loader = FileSystemLoader(template_folder),   trim_blocks=True, lstrip_blocks=True)
        template_file_name = f"{template_name}{jinja_extension}"
        template = env.get_template(template_file_name)
        sql_filename = f"{template_name}{yaml_extension}"
        output_filename_path = f"{output_path_folder}\{sql_filename}"
        write_file_local(output_filename_path, template.render(script_dict))

    except Exception as e1:
        print(e1)
        print("File Name : {}".format(template_file_name))
        pass
    
    dt2 = datetime.now()
    print(dt2-dt1)
```

```python
print("Begin of Processing")

gen_script_using_template()

print("End of Processing")
```
