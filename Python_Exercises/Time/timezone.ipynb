```python
from datetime import datetime, timezone
import pendulum
str_datetime = "2021-07-07 12:56:01"
obj_datetime = datetime.strptime(str_datetime, '%Y-%m-%d %H:%M:%S')
print(obj_datetime)
print(type(obj_datetime))
current_timezone = pendulum.timezone("US/Central")
obj_datetime_tz = pendulum.from_format(str_datetime, 'YYYY-MM-DD HH:mm:ss', tz=current_timezone)
print(obj_datetime_tz)
print(type(obj_datetime_tz))
print(int(obj_datetime_tz.timestamp()) * 1000)
```

```python
from datetime import datetime, timedelta
import pendulum
current_timezone = pendulum.timezone("US/Central")
current_datetime = pendulum.now(current_timezone)
print(current_datetime)
day_of_week = int(current_datetime.strftime('%w'))
print("Day of the Week - ", day_of_week)
if day_of_week == 1:
    date_diff = 3
else:
    date_diff = 1

prev_weekday_datetime = current_datetime - timedelta(days=date_diff)
print(prev_weekday_datetime.strftime("%Y%m%d"))
print(prev_weekday_datetime.strftime("%d-%b-%Y"))
```

```python
from datetime import datetime, timedelta
import pendulum
current_timezone = pendulum.timezone("US/Central")

start_time = "2025-10-20 09:49:04.215165+00:00"
conv_start_time = datetime.strptime(start_time, '%Y-%m-%d %H:%M:%S.%f%z')
start_time_local = current_timezone.convert(conv_start_time)
print(str(start_time_local)[:26])

end_time = "2025-10-20 09:55:02.554810+00:00"
conv_end_time = datetime.strptime(end_time, '%Y-%m-%d %H:%M:%S.%f%z')
end_time_local = current_timezone.convert(conv_end_time)
print(end_time_local)

load_run_time = str(end_time_local - start_time_local)

print(load_run_time)
```
