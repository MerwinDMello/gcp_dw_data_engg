```python
# Add items to a list
lista = ['Melwin']
stringa = 'Merwin,Joyce'
lista.extend(stringa.split(','))
for item in lista:
    print(item)
print(len(lista))
```

```python
# Get exclusive items from one list when compared to another
lista = ['Melwin','Merwin','Jyothi','Joyce']
listb = ['Melwin','Merwin']
# new_list = filter(lambda i: i not in listb,lista)
new_list = [i for i in lista if i not in listb]
print(str(new_list))
```

```python
names = ['Melwin','Merwin']
sentence = "Merwin is a boy"
if any(substring in sentence for substring in names):
    print(sentence)
else:
    print("Not present")
```

```python
from datetime import datetime as dt
from datetime import timedelta
date_fields = ['2025-03-05 04:51:32.367','2025-02-08 01:35:59.458','2025-01-17 11:47:39.000']
print(date_fields)
field_time_value = []
new_date_fields = []
for date_field in date_fields:
    field_time_value = dt.strptime(date_field,'%Y-%m-%d %H:%M:%S.%f')
    microseconds = field_time_value.microsecond
    if microseconds > 0:
        field_time_value = field_time_value.replace(microsecond=0) + timedelta(seconds=1)
    new_date_fields.append(field_time_value.strftime('%Y-%m-%d %H:%M:%S'))
print(new_date_fields)
```
