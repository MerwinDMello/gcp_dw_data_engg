```python
dt1 = datetime.now()
input_path = r"..\Files\Input_Files\File_Without_RowNo.txt"
output_path = "..\Files\Output_Files"
output_file = "File_With_RowNo.txt"
output_full_path = f"{output_path}\{output_file}"
contents = futil.read_file(input_path)
futil.make_directory_if_not_exists(output_path)
delimiter = ','
quote_char = '"'

numbered_list_of_lines = []
lines = contents.split("\n")
for index, line in enumerate(lines):
        if index != 0 :
            numbered_lines = line.strip() + delimiter + quote_char + "{0:0>8}".format(index) + quote_char 
        else :
            numbered_lines = line.strip()
        numbered_list_of_lines.append(numbered_lines)
contents = '\n'.join(numbered_list_of_lines)

futil.write_file(output_full_path, contents)
dt2 = datetime.now()
print(dt2-dt1)
```
