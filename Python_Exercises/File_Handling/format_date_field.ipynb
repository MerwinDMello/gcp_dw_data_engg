```python
from datetime import datetime
import os
import sys
```

```python
# script_dir = os.path.dirname(__file__)
script_dir = os.path.abspath('')
util_dir = os.path.join(script_dir, '..', 'Common_Utils')
sys.path.append(util_dir)
import file_utilities as futil
```

```python
dt1 = datetime.now()
delimiter = ','
input_path = r"..\Files\Input_Files\Date_Field_Without_Hyphen.txt"
output_path = "..\Files\Output_Files"
output_file = "Date_Field_With_Hyphen.txt"
output_full_path = f"{output_path}\{output_file}"
contents = futil.read_file(input_path)
futil.make_directory_if_not_exists(output_path)

field_number = 1
header_rec = True
delimiter = ','
lines = contents.split("\n")
file_lines = []
for index, line in enumerate(lines):
    if line.strip() != '' and (index == 0 and header_rec == True) :
        file_lines.append(line)
    if line.strip() != '' and (index != 0 and header_rec == True) :
        fields = line.split(",",field_number)
        date_field_in_numbers = fields[field_number - 1]
        fields[field_number - 1] = datetime.strptime(date_field_in_numbers,'%Y%m%d').strftime('%Y-%m-%d')
        file_lines.append((delimiter).join(fields))
contents = ('\n').join(file_lines)

futil.write_file(output_full_path, contents)
dt2 = datetime.now()
print(dt2-dt1)
```
